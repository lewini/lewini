```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_parquet('data/fhv_tripdata_2021-01.parquet')
len(df)
```

```python
df['duration'] = df.dropOff_datetime - df.pickup_datetime
```

```python
df
```

```python
df['duration'] = df.duration.apply(lambda td: td.total_seconds()/60)
```

```python
df.duration.mean()
```

```python
dff = df[(df.duration >= 1) & (df.duration <=60)]
```

```python
len(df) - len(dff)
```

```python
dff
```

```python
dropped = df[~df.isin(dff)].dropna(how='all')
```

```python
len(dropped)
```

```python
dropped
```

```python
dff.PUlocationID = dff.PUlocationID.fillna(-1)
```

```python
dff.DOlocationID = dff.DOlocationID.fillna(-1)
```

```python
dff
```

```python
dff.PUlocationID.value_counts(normalize=True)
```

```python
dff.DOlocationID.value_counts(normalize=True)
```

```python
dfx = dff[['PUlocationID','DOlocationID']]
```

```python
dfx
```

```python
traindict = dfx.to_dict(orient='records')
```

```python
xtrain = DictVectorizer().fit_transform(traindict)
```

```python
xtrain
```

```python
xtrain.get_shape()
```

```python
dfy = dff[['duration']]
```

```python
ytrain = dff.duration
```

```python
ytrain
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
ypred = lr.predict(xtrain)
```

```python
ypred
```

```python
mean_squared_error(ytrain, ypred, squared=False)
```
